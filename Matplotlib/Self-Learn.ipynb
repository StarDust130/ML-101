```python
# Import 
import matplotlib.pyplot as plt
```

### Line Plot ðŸ“ˆ

```python
# Data
days = [1, 2, 3, 4, 5 ,6] # X axis
sales = [100, 120, 90, 140, 80, 200] # Y axis

# Create the plot
plt.plot(days, sales , marker='o' , )

# Add title and labels
plt.title('Daily Sales Over a Week')
plt.xlabel('Days')
plt.ylabel('Sales ($)')

# Show the plot
plt.show()
```

### ðŸ”´ 2. SCATTER PLOT

```python
# Data for plotting
hours = [1, 2, 3, 4, 5, 6]
scores = [40, 50, 55, 65, 80, 85]

# Create the plot
plt.scatter(hours, scores, color='red', marker='x')

# Add title and labels
plt.title('Study Hours vs. Scores')
plt.xlabel('Hours Studied')
plt.ylabel('Scores')

# Show the plot
plt.show()
```
