```python
# Import 
import matplotlib.pyplot as plt
```

### Line Plot ðŸ“ˆ

```python
# Data
days = [1, 2, 3, 4, 5 ,6] # X axis
sales = [100, 120, 90, 140, 80, 200] # Y axis

# Create the plot
plt.plot(days, sales , marker='o' , )

# Add title and labels
plt.title('Daily Sales Over a Week')
plt.xlabel('Days')
plt.ylabel('Sales ($)')

# Show the plot
plt.show()
```

### ðŸ”´ 2. SCATTER PLOT

```python
# Data for plotting
hours = [1, 2, 3, 4, 5, 6]
scores = [40, 50, 55, 65, 80, 85]

# Create the plot
plt.scatter(hours, scores, color='red', marker='x')

# Add title and labels
plt.title('Study Hours vs. Scores')
plt.xlabel('Hours Studied')
plt.ylabel('Scores')

# Show the plot
plt.show()
```

### ðŸŸ¡ 3. HISTOGRAM

```python
# Data for plotting
ages = [22, 25, 30, 30, 32, 40, 41, 42, 45, 50, 51, 52]

# Create the histogram
plt.hist(ages, bins=5,  edgecolor='black')

# Add title and labels
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

sales = [120, 150, 180, 130, 170, 200, 220]
days = [1, 2, 3, 4, 5, 6, 7]

# Line plot
plt.plot(days, sales)
plt.title("Sales Trend")
plt.show()

# Scatter plot
plt.scatter(days, sales)
plt.title("Sales Scatter")
plt.show()

# Histogram (distribution)
plt.hist(sales, bins=5)
plt.title("Sales Distribution")
plt.show()
```
