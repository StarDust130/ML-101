```python
X = [1, 2, 3, 4, 5]
Y = [10, 20, 30, 40, 50]

plt.plot(X, Y);
```

```python
# 1st method
fig = plt.figure() # create a figure 
ax = fig.add_subplot() # add some axes
ax.plot(X, Y); # plot on the axes
```

```python
# 2nd method
fig = plt.figure() # create a figure
ax = fig.add_axes([1,1,1,1])
ax.plot(X,Y) # add data to axes
plt.show()
```

```python
# 3 method (recommended )
fig, ax = plt.subplots() # create figure and axes
ax.plot(X, [50,100,200,350 , 500]); # plot on the axes
plt.show()
```

## Matplotlib example workflow

```python
# 0. import matplotlib and get it ready for plotting in Jupyter
%matplotlib inline
import matplotlib.pyplot as plt

# 1. preare data
x = [1, 2, 3, 4]
y = [11 , 22, 33, 44]

#2. setup plot
fig , ax = plt.subplots(figsize=(10,10)) # (width, height)

# 3. plot the data
ax.plot(x, y)

# 4. customize the plot
ax.set(title="Simple Plot", 
       xlabel="x-axis", 
       ylabel="y-axis")

# 5. save and show (you save the figure first then show it)
fig.savefig("sample-plot.png")

plt.show() # shows the plot 





```

## Making figures with NumPy arrays

We want:
- Line plot
- Scatter plot
- Bar plot
- Histogram
- Subplots
- Pie chart


```python
#  Create some data
x = np.linspace(0, 10, 100)
y = np.sin(x)

#  Plot the data and create a line plot
fig, ax = plt.subplots(figsize=(10, 6))

ax.plot(x, y);
ax.set(title="Line Plot",
       xlabel="X-axis",
       ylabel="Y-axis")

plt.show()
```

```python
# Use same data to make a scatter plot
fig, ax = plt.subplots(figsize=(10, 6))
ax.scatter(x, y);
ax.set(title="Scatter Plot",
       xlabel="X-axis",
       ylabel="Y-axis")

plt.show()
```

```python
# Make a plot from dictionary (bar plot data)
nut_butter_prices = {
    "Almond Butter": 10,
    "Peanut Butter": 8,
    "Cashew Butter": 12
    }

fig, ax = plt.subplots(figsize=(10, 6))
ax.bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax.set(title="Nut Butter Prices (BAR PLOT)",
       xlabel="Type",
       ylabel="Price ($)")

plt.show()
```

```python
# This code makes a histogram chart  to show how numbers are spread out
# First, we create 1000 random numbers from a normal bell curve 
x = np.random.randn(1000)

# Next, we set up a big picture space for the chart 
fig, ax = plt.subplots(figsize=(10, 6))

# Now, we draw the histogram with 30 bars to group the numbers 
ax.hist(x, bins=30);

# Add a title and labels to make it clear 
ax.set(title="Histogram Plot",
       xlabel="Value",
       ylabel="Frequency")

# Finally, show the cool chart 
plt.show()




```

```python
# Make some data
x = np.random.randn(1000)
# Set up the figure and axiss
fig, ax = plt.subplots(figsize=(10, 6))
# Create a histogram from the data
ax.hist(x, bins=30);
ax.set(title="Histogram Plot",
       xlabel="Value",
       ylabel="Frequency")

plt.show()
```

### Two options for subplots


```python
# 1.) Subplots option 1
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2, ncols=2, figsize=(10, 5))

# Plot to each different axis
ax1.plot(x, x/2);
ax1.set(title="Axis 1")

ax2.scatter(np.random.randn(10), np.random.randn(10));
ax2.set(title="Axis 2")

ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax3.set(title="Axis 3")

ax4.hist(x, bins=30);
ax4.set(title="Axis 4")

plt.tight_layout(); # adjust the spacing between subplots
plt.show()
```
