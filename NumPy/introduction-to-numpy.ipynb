## Dot product example (nut butter sales)

```python
np.random.seed(0)
# Numbers of jars sold 
sales_amounts = np.random.randint(20, size=(5, 3)) # 5 sales, 3 products
sales_amounts
```

```python
# Create weekly_sales Dataframe 
weekly_sales = pd.DataFrame(sales_amounts, 
                            index=["Mon", "Tue", "Wed", "Thu", "Fri"], 
                            columns=["Almond Butter", "Peanut Butter", "Cashew Butter"])
weekly_sales
```

```python
# Create prices array
prices = np.array([10, 8, 12])  # prices for each product
prices
```

```python
# Create butter_prices DataFrame
butter_prices = pd.DataFrame(prices.reshape(1,3), 
                             index=["Price"], 
                             columns=["Almond Butter", "Peanut Butter", "Cashew Butter"])
butter_prices
```

```python
total_sales = prices.dot(sales_amounts.T) # T means align the shapes for dot product
total_sales
```

```python
# Create  daily_sales
butter_prices.shape , weekly_sales.shape
```

```python
daily_sales = butter_prices.dot(weekly_sales.T)
daily_sales
```
